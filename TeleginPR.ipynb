```python
while True:
    print("\n--- Калькулятор ---")

    try:
        num1 = float(input("Введите первое число: "))

        num2 = float(input("Введите второе число: "))

        operator = input("Введите оператор (+, -, *, /) или 'q' для выхода: ")

        if operator.lower() == 'q':
            print("Выход из программы...")
            break

        if operator == "+":
            result = num1 + num2
            print(f"{num1} + {num2} = {result}")
        elif operator == "-":
            result = num1 - num2
            print(f"{num1} - {num2} = {result}")
        elif operator == "*":
            result = num1 * num2
            print(f"{num1} * {num2} = {result}")
        elif operator == "/":
            if num2 != 0:
                result = num1 / num2
                print(f"{num1} / {num2} = {result}")
            else:
                print("Ошибка: деление на ноль невозможно!")
        else:
            print("Ошибка: неверный оператор!")

    except ValueError:
        print("Ошибка: введите корректные числа!")
```

```python
import random
import string

def generate_password(length, use_letters=True, use_digits=True, use_special=True, count=1):

    chars = ""

    if use_letters:
        chars += string.ascii_letters
    if use_digits:
        chars += string.digits
    if use_special:
        chars += string.punctuation

    if not chars:
        return ["Ошибка: выберите хотя бы один тип символов!"]

    passwords = []
    for _ in range(count):
        password = ''.join(random.choice(chars) for _ in range(length))
        passwords.append(password)

    return passwords

def check_password_strength(password):
    """
    Проверяет сложность пароля и возвращает оценку и рекомендации

    Возвращает: (оценка, список рекомендаций)
    оценка: "Очень слабый", "Слабый", "Средний", "Надежный", "Очень надежный"
    """
    score = 0
    recommendations = []

    length = len(password)
    if length < 8:
        recommendations.append("Увеличьте длину пароля до 8+ символов")
    elif length < 10:
        score += 1
    elif length < 12:
        score += 2
    else:
        score += 3

    if any(c.isupper() for c in password):
        score += 1
    else:
        recommendations.append("Добавьте буквы верхнего регистра")

    if any(c.islower() for c in password):
        score += 1
    else:
        recommendations.append("Добавьте буквы нижнего регистра")

    if any(c.isdigit() for c in password):
        score += 1
    else:
        recommendations.append("Добавьте цифры")

    if any(c in string.punctuation for c in password):
        score += 1
    else:
        recommendations.append("Добавьте специальные символы")

    if score <= 2:
        strength = "Очень слабый"
    elif score <= 3:
        strength = "Слабый"
    elif score <= 4:
        strength = "Средний"
    elif score <= 5:
        strength = "Надежный"
    else:
        strength = "Очень надежный"

    return strength, recommendations

def main():
    """Основная функция программы"""
    print("=" * 50)
    print(" ГЕНЕРАТОР БЕЗОПАСНЫХ ПАРОЛЕЙ")
    print("=" * 50)

    while True:
        print("\n--- Настройки генерации ---")

        try:
            length = int(input("Длина пароля (минимум 6): "))
            if length < 6:
                print("Длина пароля должна быть не менее 6 символов. Установлено значение 6.")
                length = 6
        except ValueError:
            print("Ошибка: введите число. Установлена длина 8.")
            length = 8

        print("\nВыберите типы символов для пароля:")
        use_letters = input("Использовать буквы? (да/нет): ").lower() in ['да', 'д', 'yes', 'y', '']
        use_digits = input("Использовать цифры? (да/нет): ").lower() in ['да', 'д', 'yes', 'y', '']
        use_special = input("Использовать специальные символы? (да/нет): ").lower() in ['да', 'д', 'yes', 'y', '']
        if not (use_letters or use_digits or use_special):
            print("Ошибка: выберите хотя бы один тип символов!")
            continue

        try:
            count = int(input("\nСколько паролей сгенерировать? (по умолчанию 1): ") or "1")
            if count < 1:
                count = 1
        except ValueError:
            print("Ошибка: введите число. Сгенерирован 1 пароль.")
            count = 1

        print(f"\n{'='*50}")
        print(f"ГЕНЕРАЦИЯ {count} ПАРОЛЕЙ (длина: {length})")
        print(f"{'='*50}")

        passwords = generate_password(length, use_letters, use_digits, use_special, count)

        for i, password in enumerate(passwords, 1):
            print(f"\nПароль {i}: {password}")

            strength, recommendations = check_password_strength(password)
            print(f"Сложность: {strength}")

            if recommendations:
                print("Рекомендации:")
                for rec in recommendations[:3]:
                    print(f"  • {rec}")

        again = input("\nСгенерировать еще пароли? (да/нет): ").lower()
        if again not in ['да', 'д', 'yes', 'y', '']:
            print("\nСпасибо за использование генератора паролей!")
            break

if __name__ == "__main__":
    main()
```

```python
from datetime import datetime
from collections import defaultdict


menu = {}
orders = []

def add_dish_to_menu(dish_id, name, price, category, prep_time):

    menu[dish_id] = {
        "name": name,
        "price": price,
        "category": category,
        "prep_time": prep_time
    }
    print(f"Блюдо '{name}' добавлено в меню.")

def create_order(table_number, dish_ids):

    order_items = []
    for dish_id in dish_ids:
        if dish_id in menu:
            order_items.append(menu[dish_id])
        else:
            print(f"Предупреждение: Блюдо с ID {dish_id} не найдено в меню и не будет добавлено в заказ.")

    if not order_items:
        print("Заказ не может быть создан, так как не содержит валидных блюд.")
        return None

    order = {
        "order_id": len(orders) + 1,
        "table": table_number,
        "items": order_items,
        "status": "принят",
        "timestamp": datetime.now().strftime("%Y-%m-%d %H:%M:%S")
    }
    orders.append(order)
    print(f"Заказ №{order['order_id']} для стола {table_number} создан.")
    return order

def calculate_order_cost(order_id):

    for order in orders:
        if order["order_id"] == order_id:
            total = sum(item["price"] for item in order["items"])
            return total
    print(f"Заказ с ID {order_id} не найден.")
    return 0

def get_popular_dishes(limit=3):

    dish_count = defaultdict(int)
    for order in orders:
        for item in order["items"]:
            dish_count[item["name"]] += 1


    sorted_dishes = sorted(dish_count.items(), key=lambda x: x[1], reverse=True)
    return dict(sorted_dishes[:limit])

def revenue_by_category():

    category_revenue = defaultdict(float)
    for order in orders:
        for item in order["items"]:
            category_revenue[item["category"]] += item["price"]

    return dict(category_revenue)


if __name__ == "__main__":
    print("--- Система кафе ---")

    add_dish_to_menu(1, "Пицца Маргарита", 450, "Пицца", 15)
    add_dish_to_menu(2, "Паста Карбонара", 380, "Паста", 12)
    add_dish_to_menu(3, "Салат Цезарь", 320, "Салаты", 8)
    add_dish_to_menu(4, "Чай", 150, "Напитки", 3)
    add_dish_to_menu(5, "Кофе", 200, "Напитки", 5)


    create_order(5, [1, 3, 4])
    create_order(3, [2, 5])
    create_order(5, [1, 4])

    cost = calculate_order_cost(1)
    print(f"Стоимость заказа №1: {cost} руб.")

    popular = get_popular_dishes(3)
    print("Самые популярные блюда:", popular)

    revenue = revenue_by_category()
    print("Выручка по категориям:", revenue)
```

```python
from statistics import mean

students = {}
grades = []

def add_student(student_id, full_name, group, contacts):
    students[student_id] = {
        "name": full_name,
        "group": group,
        "contacts": contacts
    }
    print(f"Студент {full_name} (ID: {student_id}) добавлен.")

def add_grade(student_id, subject, grade, date):
    if student_id not in students:
        print(f"Ошибка: Студент с ID {student_id} не найден.")
        return

    grades.append({
        "student_id": student_id,
        "subject": subject,
        "grade": grade,
        "date": date
    })
    print(f"Оценка {grade} по предмету '{subject}' добавлена для студента ID {student_id}.")

def calculate_student_avg(student_id):
    student_grades = [g["grade"] for g in grades if g["student_id"] == student_id]
    if not student_grades:
        return 0.0
    return round(mean(student_grades), 2)

def find_low_performance_students(threshold=3.0):
    low_performers = {}
    for student_id in students:
        avg = calculate_student_avg(student_id)
        if avg < threshold:
            low_performers[student_id] = {
                "name": students[student_id]["name"],
                "average": avg
            }
    return low_performers

def get_students_rating():
    rating = []
    for student_id in students:
        avg = calculate_student_avg(student_id)
        rating.append({
            "id": student_id,
            "name": students[student_id]["name"],
            "average": avg
        })
    return sorted(rating, key=lambda x: x["average"], reverse=True)

if __name__ == "__main__":
    print("\n--- Задача 2: Успеваемость студентов ---")
    add_student(101, "Иванов Иван", "ИТ-21", "ivan@mail.ru")
    add_student(102, "Петрова Анна", "ИТ-21", "petrova@mail.ru")
    add_student(103, "Сидоров Петр", "БИО-12", "sidorov@mail.ru")

    add_grade(101, "Математика", 5, "2024-01-20")
    add_grade(101, "Физика", 4, "2024-01-21")
    add_grade(102, "Математика", 3, "2024-01-20")
    add_grade(102, "Физика", 3, "2024-01-21")
    add_grade(103, "Биология", 4, "2024-01-22")
    add_grade(103, "Химия", 5, "2024-01-23")

    avg_ivan = calculate_student_avg(101)
    print(f"Средний балл Иванова: {avg_ivan}")

    low_students = find_low_performance_students(4.0)
    print("Студенты со средним баллом ниже 4.0:", low_students)

    rating = get_students_rating()
    print("Рейтинг студентов:")
    for idx, student in enumerate(rating, 1):
        print(f"  {idx}. {student['name']} - {student['average']}")
```

```python
import re

data_text = """
Посетите наши страницы:
Email: info@example.com (главный), support-team@corp.net (поддержка)
Дата: 25-05-2023, 01/01/2024, 12.02.2025.
Коды товаров: ID_PROD-1001, ITEM-20-A, ID-300-B.
Суммы: $150.50, 20 EUR, 500 RUB, £75.
"""

print("--- Email адреса ---")
email_pattern = r'[a-zA-Z0-9_.-]+@[a-zA-Z0-9-]+\.[a-zA-Z]{2,}'
emails = re.findall(email_pattern, data_text)
print(emails)

print("\n--- Цифровые части кодов товаров ---")
code_pattern = r'(?:ID_PROD-|ID-)(\d+)'
codes = re.findall(code_pattern, data_text)
print(codes)


print("\n--- Замена формата даты ---")
date_pattern = r'(\d{2})\.(\d{2})\.(\d{4})'

updated_text = re.sub(date_pattern, r'\3-\2-\1', data_text)
print("Текст после замены формата даты:")
print(updated_text)

print("\n--- Задание 4: Суммы в EUR или RUB ---")
eur_rub_pattern = r'\b\d+(?:\.\d+)?\s+(EUR|RUB)\b'
eur_rub_amounts = re.findall(eur_rub_pattern, data_text)

full_matches = re.findall(r'\b\d+(?:\.\d+)?\s+(?:EUR|RUB)\b', data_text)

print("Найденные валюты:", eur_rub_amounts)
print("Полные суммы:", full_matches)
```

```python
import re

def validate_inn():
    while True:
        inn = input("Введите ИНН (12 цифр): ").strip()

        if re.fullmatch(r'\d{12}', inn):
            print(f"ИНН принят: {inn}")
            return inn
        else:
            print("Ошибка! ИНН должен состоять ровно из 12 цифр.")

print("--- Проверка ИНН ---")
user_inn = validate_inn()

def validate_price():
    while True:
        price_input = input("Введите цену товара (положительное число до 1 000 000): ").strip()

        try:
            price = float(price_input)

            if price <= 0:
                print("Ошибка! Цена должна быть положительным числом.")
            elif price > 1_000_000:
                print("Ошибка! Цена не должна превышать 1 000 000.")
            elif price != price:
                print("Ошибка! Введено некорректное число.")
            else:
                print(f"Цена принята: {price:.2f}")
                return price

        except ValueError:
            print("Ошибка! Пожалуйста, введите корректное число (например: 99.99 или 100).")

print("\n--- Валидация цены ---")
user_price = validate_price()

def validate_password():
    while True:
        password = input("Введите пароль (8-20 символов, минимум 1 заглавная буква и 1 цифра): ")

        if len(password) < 8:
            print("Ошибка! Пароль слишком короткий. Минимум 8 символов.")
            continue
        elif len(password) > 20:
            print("Ошибка! Пароль слишком длинный. Максимум 20 символов.")
            continue

        if not re.search(r'[A-Z]', password):
            print("Ошибка! Пароль должен содержать хотя бы одну заглавную букву (A-Z).")
            continue

        if not re.search(r'\d', password):
            print("Ошибка! Пароль должен содержать хотя бы одну цифру (0-9).")
            continue

        if not re.fullmatch(r'[A-Za-z0-9]+', password):
            print("Ошибка! Пароль может содержать только буквы латинского алфавита и цифры.")
            continue

        print("Пароль принят")
        return password

print("\n--- Валидация пароля ---")
user_password = validate_password()
```
